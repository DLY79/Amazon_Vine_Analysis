```python
import pandas as pd
```

```python
vine_df= pd.read_csv('vine_table.csv')
```

```python
vine2_df=vine_df[vine_df['total_votes']>= 20]
```

```python
vine2_df.describe()
```

```python
helpful_votes_df=vine2_df[(vine2_df['helpful_votes'])/(vine2_df['total_votes'])>= 0.5]
```

```python
helpful_votes_df
```

```python
yesVine_df = helpful_votes_df[helpful_votes_df['vine']== 'Y']
```

```python
yesVine_df
```

```python
nonVine_df = helpful_votes_df[helpful_votes_df['vine']== 'N']
```

```python
nonVine_df
```

```python
yesVine_df.describe()
```

```python
nonVine_df
```

```python
results_df=pd.concat([nonVine_df, yesVine_df], axis=0, join="inner")
```

```python
results2_df=results_df.groupby(['vine'])
```

```python
results_df
```

```python
total_reviews = results_df['review_id'].count()
total_reviews
```

```python
total_5_star = results_df[results_df['star_rating'] ==5.0].count()['star_rating']
total_5_star
```

```python
unpaid_5_star = (nonVine_df[nonVine_df['star_rating'] ==5.0].count()['star_rating'])/(nonVine_df['review_id'].count()) * 100
print("Unpaid Vine Reviews: ", unpaid_5_star, '%')
```

```python
paid_5_star = (yesVine_df[yesVine_df['star_rating'] ==5.0].count()['star_rating'])/(yesVine_df['review_id'].count()) * 100
print("Paid Vine Reviews: ", paid_5_star, '%')
```

```python
summary_df = pd.DataFrame ({
    "total reviews" : [total_reviews],
     "total 5 stars" : [total_5_star],
     "Unpaid 5 stars %" : [unpaid_5_star],
     "paid 5 stars %" : [paid_5_star],
    
})
```
